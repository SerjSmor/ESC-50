```python
# def tran_and_fit_fold(meta, fold_number)
def train_test_fold(fold):
    full_train = meta[meta['fold'] != fold]
    full_test = meta[meta['fold'] == fold]

    X_train = full_train['filename']
    X_test = full_test['filename']

    y_train = full_train['target']
    y_test = full_test['target']

    # for each file in training we need to get it's features 
    X_train_vectors = []
    for index, filename in X_train.items():
        csv_file_path = os.path.join(csv_folder, filename)
        features = loadtxt(csv_file_path, delimiter=',')
        X_train_vectors.append(features)

    X_test_vectors = []
    for index, filename in X_test.items():
        csv_file_path = os.path.join(csv_folder, filename)
        features = loadtxt(csv_file_path, delimiter=',')
        X_test_vectors.append(features)


    regressor = RandomForestClassifier(n_estimators=200, random_state=0)
    regressor.fit(X_train_vectors, y_train)
    y_pred = regressor.predict(X_test_vectors)
    print(len(y_pred), len(y_test))
    score = accuracy_score(y_test, y_pred)
    print(score)
```

```python

for i in range(1,6):
    train_test_fold(i)
```
